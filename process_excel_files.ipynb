# To make synonym dictionary and remove duplicate

```python
import os
import pandas as pd
import time
import numpy as np
```

```python
file_list = os.listdir('.')
print(file_list)
```

```python
fname = 'synonymous gene ALL BACILLUS.xlsx'
df = pd.read_excel(fname, usecols=['word', 'syn'])
print(df.columns)
# print(df.head)
synonym_dict = {}
for index, row in df.iterrows():
    root = row['word'] 
    syn = row['syn']
    #print(index, row['word'], row['syn'])
    if  type(syn) != float: # to check for nan values
        if root not in synonym_dict:
            synonym_dict[root] = [syn]
        else:
            synonym_dict[root].append(syn)
        if syn not in synonym_dict:
            synonym_dict[syn] = [root]
        else:
            synonym_dict[syn].append(root)
    else:
        print('\t At  {}th row,  NULL found'.format(index))
print(synonym_dict)
```

```python
fname = 'bacillus_no_other_mol_excluded.csv'
df2 = pd.read_csv(fname)
print(df2.columns)
print(df2.head)
```

```python
seen_dict = {}
for index, row in df2.iterrows():
    key = row['Source']
    value = row['Target']
    if key not in seen_dict:
        seen_dict[key] = [value]
    else:
        seen_dict[key].append(value)
        print(' \t Same key found at index : {} '.format(index))
```

```python
key_list = list(seen_dict.keys())
key_list.sort()
for key in key_list:
    print(key, ' : ' , seen_dict[key])
```

```python
# save seen_dict
import json
fname_out = 'seen_dict.json'
fout = open(fname_out, 'w')
json.dump( seen_dict, fout, indent =4)
fout.close()
```

```python
# save synonym_dict
fname_out = 'synonym_dict.json'
fout = open(fname_out, 'w')
json.dump(synonym_dict, fout, indent =4)
fout.close()
```

```python
# extarct according to Rule 1 : same key but value are synonym
for src in seen_dict:
    dest_list = seen_dict[src]
    for dest in dest_list:
        if dest not in synonym_dict:
            continue
        else:
            #print('Checking the synonym dict')
            synonym_list = set(synonym_dict[dest])
            intersection_list = list((set(dest_list) - set([dest])) & set(synonym_list))
            if len(intersection_list) >1:
                print(' Src : {}, dest: {}, intersection: {}'.format(src, dest, intersection_list))
# sigA ->  dnaK, sigA -> dnaA, sigA -> dnaN
```

```python
# Rule 2:  src synonym , destination same 
#src1 -> []
#src2 -> []
src_list = list(seen_dict.keys())
for src in src_list:
    if src not in synonym_dict:
        continue
    else:
        synonym_list = synonym_dict[src]
        modified_src_set = set(src_list) - set([src])
        intersection_list = list( modified_src_set & set(synonym_list))
        if len(intersection_list) >1:
            print(' src : {} , intersection list : {}'.fromat(src, intersection_list))
```
